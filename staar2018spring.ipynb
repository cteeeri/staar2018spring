##### English 1 #####

```python
df_eng1copy = df_eng1.iloc[:,[0,5]].copy()
df_eng1copy.rename(index=str, columns={'STAAR - Approaches - %':'Eng1'}, inplace=True)
df_eng1copy.head(3)
```

##### Eng 2 #####

```python
df_eng2copy = df_eng2.iloc[:,[0,5]].copy()
df_eng2copy.rename(index=str, columns={'STAAR - Approaches - %':'Eng2'}, inplace=True)
df_eng2copy.head(3)
```

##### US History #####

```python
df_ushistoryCopy = df_ushistory.iloc[:,[0,5]].copy()
df_ushistoryCopy.rename(index=str, columns={'STAAR - Approaches - %':'History'}, inplace=True)
df_ushistoryCopy.head(3)
```

##### Biology #####

```python
df_biologyCopy = df_biology.iloc[:,[0,5]].copy()
df_biologyCopy.rename(index=str, columns={'STAAR - Approaches - %':'Biology'}, inplace=True)
df_biologyCopy.head(3)
```

#### Combine #####

```python
met_eoc = df_alg1copy.merge(df_eng1copy, how='outer', on='Group')
met_eoc = met_eoc.merge(df_eng2copy, how='outer', on='Group')
met_eoc = met_eoc.merge(df_ushistoryCopy, how='outer', on='Group')
met_eoc = met_eoc.merge(df_biologyCopy, how='outer', on='Group')
```

```python
met_eoccopy = met_eoc.copy()
met_eoccopy['Grade'] = 'EOC'
met_eoccopy
```

```python
met_eoccopy.drop_duplicates(inplace=True)
met_eoccopy.reset_index(drop=True,inplace=True)
```

```python
met_eoccopy.drop([7, 9, 11, 13, 16, 18, 20, 22, 24], inplace=True)
met_eoccopy.reset_index(drop=True, inplace=True)
```

```python
met_eoccopy.replace('UPLIFT HAMPTON PREPARATORY H S','HAMPTON', inplace=True)
met_eoccopy.replace('UPLIFT HEIGHTS PREPARATORY H S','HEIGHTS', inplace=True)
met_eoccopy.replace('UPLIFT LUNA PREPARATORY H S','LUNA', inplace=True)
met_eoccopy.replace('UPLIFT PEAK PREPARATORY H S','PEAK', inplace=True)
met_eoccopy.replace('UPLIFT WILLIAMS PREPARATORY H S','WILLIAMS', inplace=True)
met_eoccopy.replace('UPLIFT EDUCATION-NORTH HILLS PREP H S','NORTH HILLS', inplace=True)
met_eoccopy.replace('UPLIFT MIGHTY H S','MIGHTY', inplace=True)
met_eoccopy.replace('UPLIFT SUMMIT INTERNATIONAL H S','SUMMIT', inplace=True)
met_eoccopy.replace('UPLIFT INFINITY H S','INFINITY', inplace=True)
met_eoccopy.replace('UPLIFT GRAND H S', 'GRAND', inplace=True)
```

### Combine STAAR38 and EOC ###

```python
met_staar38['Alg1'] = np.nan
met_staar38['Eng1'] = np.nan
met_staar38['Eng2'] = np.nan
met_staar38['History'] = np.nan
met_staar38['Biology'] = np.nan
met_staar38
```

### Rearrange row order to ease cut and paste to designated spreadsheet ###

```python
pd.Series(met.index).values
```

```python
met = met.reindex(['State','ARLINGTON ISD', 'DALLAS ISD (057905)', 'FORT WORTH ISD', 'GRAND PRAIRIE ISD',
             'IRVING ISD','UPLIFT EDUCATION','UPLIFT GRADUS PREPARATORY','GRAND','HAMPTON', 
             'HEIGHTS', 'INFINITY','LUNA',  'UPLIFT MERIDIAN SCHOOL','MIGHTY',
             'NORTH HILLS', 'PEAK', 'UPLIFT PINNACLE PREPARATORY PRI','SUMMIT',
             'UPLIFT TRIUMPH PREPARATORY SCHOOL', 'WILLIAMS'])
```

```python
met
```

### Rearrange column order ###

```python
met = met.swaplevel(axis=1)
met.columns
met2 = met.copy()
```

```python
met2.columns.get_level_values(0)
met2.columns.get_level_values(1)
```

```python
grades_level = pd.Categorical(met2.columns.get_level_values(0), 
                              categories=[3, 4, 5, 6, 7, 8, 'EOC'],
                             ordered=True)
subjects_level = pd.Categorical(met2.columns.get_level_values(1),
                               categories=['Math', 'Reading', 'Sc', 'SS', 'Writing','Alg1', 'Eng1', 'Eng2', 'History','Biology'],
                               ordered=True)
met2.columns = pd.MultiIndex.from_arrays([grades_level, subjects_level])
met2 = met2.sort_index(axis=1, level=[0,1])
met2;
```

```python
met2.columns.names = ['Subjects', 'Grades']
met2
```
